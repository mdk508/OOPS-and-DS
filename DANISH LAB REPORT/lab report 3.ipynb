```python
x = ["a", ["bb", ["ccc", "ddd"], "ee", "ff"], "g", ["hh", "ii"], "j"]
```

```python
x = ["a", ["bb", ["ccc", "ddd"], "ee", "ff"], "g", ["hh", "ii"], "j"]

print(x[0])  # "a"
print(x[1][0])  # "bb"
print(x[1][1][0])  # "ccc"
print(x[1][1][1])  # "ddd"
print(x[1][2])  # "ee"
print(x[1][3])  # "ff"
print(x[2])  # "g"
print(x[3][0])  # "hh"
print(x[3][1])  # "ii"
print(x[4])  # "j"
```
